```python
import pandas as pd
```

```python
dataset=pd.read_csv("Social_Network_Ads.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset
```

```python
dataset=dataset.drop('User ID',axis=1)
```

```python
dataset['Purchased'].value_counts()
```

```python
independent=dataset[['Age', 'EstimatedSalary', 'Gender_Male']]
dependent=dataset[['Purchased']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.tree import DecisionTreeClassifier

from sklearn.model_selection import GridSearchCV
param_grid={'criterion':['gini','entropy','log_loss'],
            'max_features':['auto','sqrt','log2'],
            'splitter':['best','random']}

grid=GridSearchCV(DecisionTreeClassifier(),param_grid,refit=True,verbose=2,cv=5,n_jobs=1,scoring='f1_weighted')
grid=grid.fit(x_train,y_train)
```

```python
re=grid.cv_results_
table=pd.DataFrame.from_dict(re)
grid_pred=grid.predict(x_test)
```

```python
table
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,grid_pred)
```

```python
cm
```

```python
from sklearn.metrics import classification_report
clf_report=classification_report(y_test,grid_pred)
print(clf_report)
```

```python
from sklearn.metrics import f1_score
f1_macro=f1_score(y_test,grid_pred,average='weighted')
print('The f1 macro value from best parameters{}:'.format(grid.best_params_),f1_macro)
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test,grid.predict_proba(x_test)[:,1])
```
